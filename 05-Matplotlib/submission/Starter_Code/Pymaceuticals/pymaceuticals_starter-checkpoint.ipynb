```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study the data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#Display the mouse metadata
mouse_metadata.head()
```

```python
#Display the study results
study_results.head()
```

```python
# Combine the data into a single DataFrame
merge_df=pd.merge(mouse_metadata, study_results, on="Mouse ID")


#Display the data table for preview
merge_df.head()
merge_df.info()
```

```python
# Checking the number of mice
# Unique Mouse IDs.
print(merge_df['Mouse ID'].nunique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
merge_df.groupby(["Mouse ID", "Timepoint"]).size().sort_values(ascending=False)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mask=merge_df["Mouse ID"]=="g989"
merge_df.loc[mask]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mask=merge_df["Mouse ID"]!="g989"
clean_df = merge_df.loc[mask].reset_index(drop=True)
clean_df.head()
```

```python
# Checking the number of unique mice IDs in the clean DataFrame.
print(clean_df['Mouse ID'].nunique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

#Group the data by drug regimen
drug_regimen_groups = clean_df.groupby(["Drug Regimen"])

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean().map('{:.2f}'.format)
median_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median().map('{:.2f}'.format)
variance_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var().map('{:.2f}'.format)
std_tumor = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std().map('{:.2f}'.format)
sem_tumor =  clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem().map('{:.2f}'.format)
summary_stats = pd.DataFrame({"Mean of Tumor":mean_tumor,"Median of Tumor":median_tumor,"Var of Tumor":variance_tumor,"Std of Tumor":std_tumor,"SEM of Tumor": sem_tumor})
 
# Assemble the resulting series into a single summary DataFrame.
print(summary_stats)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats = clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
#Step 1: Create a DataFrame with only these four treatment regimens
treatment_list = ["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

treatment_df=clean_df[clean_df["Drug Regimen"].isin(treatment_list)]

# Step 2: Get the last (greatest) timepoint for each mouse
last_timepoints= clean_df.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Step 3: Merge the last_timepoints DataFrame with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume_df=pd.merge(treatment_df, last_timepoints, on=['Mouse ID', 'Timepoint'])

# Step 4:Calculate the final tumor volume for each mouse
final_tumor_vol = final_tumor_volume_df[['Mouse ID', 'Tumor Volume (mm3)', 'Drug Regimen']]

# Step 5: Display the final tumor volume of each mouse across four of the treatment regimens
final_tumor_vol
```

```python
# Put treatments into a list for loop (and later for plot labels)
treatment= ["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
final_tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatment:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol2 = final_tumor_vol.loc[final_tumor_vol["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    
    # Add subset final_tumor_vol_list.append(final_tumor_vol) 
    final_tumor_vol_list.append(final_tumor_vol2)
    
    quartiles = final_tumor_vol2.quantile([0.25,0.5,0.75]) 
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    mean = final_tumor_vol2.mean()
    iqr = upper_q-lower_q 
    lower_bound = lower_q - (1.5*iqr) 
    upper_bound = upper_q + (1.5*iqr) 
    
    # Determine outliers using upper and lower bounds
    outliers = final_tumor_vol2.loc[(final_tumor_vol2 < lower_bound)|(final_tumor_vol2 >upper_bound)]
    print(f'{drug}: {outliers}')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(final_tumor_vol_list, labels=treatment, flierprops=dict(marker='o', markerfacecolor='r', markersize=7))

plt.ylabel("Final Tumor Volume (mm3)")
plt.title("Final Tumor Volume for Each Treatment Drug")
plt.show()  
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Filter the data to only include the Capomulin treatment regimen
capomulin_data = clean_df[clean_df['Drug Regimen'] == 'Capomulin']

# Calculate the average tumor volume for each mouse in the Capomulin group
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
# avg_tumor_volume

# Merge the average tumor volume data with the mouse metadata to get the weight for each mouse
capomulin_data_with_weight = pd.merge(avg_tumor_volume, mouse_metadata, on='Mouse ID')

# Create the scatter plot
plt.scatter(capomulin_data_with_weight['Weight (g)'], capomulin_data_with_weight['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Avg Tumor Volume in Capomulin Regimen', fontsize = 10)
plt.grid(color="lightgrey",alpha =0.2,linestyle = "--")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model for mouse weight and average observed tumor volume for the entire Capomulin regimen
#Calculate the correlation coefficient and linear regression model

x_axis = capomulin_data_with_weight['Weight (g)']
y_axis = capomulin_data_with_weight['Tumor Volume (mm3)']
print(f"The correlation coefficient between mouse weight and average tumor volume is\{round(st.pearsonr(x_axis,y_axis)[0],2)}")

slope, intercept, rvalue, pvalue, std_err = st.linregress(x_axis,y_axis)
line_eq="y = "+str(round(slope,2))+"x + "+str(round(intercept,2))
regress_fit=slope*x_axis+intercept

# Add the linear regression line to the scatter plot
print(f"Linear regression equation: y = {slope}x + {intercept}")
```

```python
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regress_fit,'r-')
plt.title('Mouse Weight vs. Avg Tumor Volume in Capomulin Regimen', fontsize=10)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume')
plt.show()
```
